```python
paths = []

for dirname, _, filenames in os.walk('/content/drive/MyDrive/shape_figure/train'):
    for filename in filenames:
        if 'cir' in filename and filename.endswith('.jpg'):
            paths.append(os.path.join(dirname, filename))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# Assuming paths is a list of image paths
random_paths = random.sample(paths, 20)

for j, path in enumerate(random_paths):
    img = cv2.imread(path)

    # Resize the image to 312x312
    img = cv2.resize(img, (312, 312))

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    try:
        circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1, minDist=20,
                                    param1=100, param2=38,
                                    minRadius=10, maxRadius=400)
        circles = np.uint16(np.around(circles))
        lenf = len(circles[0, :])
        if lenf > 0:
            f, ax = plt.subplots(1, 1 + lenf, figsize=(3 * (1 + lenf), 3))
            ax[0].imshow(img, aspect='auto')
            ax[0].set_xticks([])
            ax[0].set_yticks([])

            for j, circle in enumerate(circles[0, :]):
                imgj = img.copy()
                cv2.circle(imgj, (circle[0], circle[1]), circle[2], (0, 165, 255), 5)
                cv2.circle(imgj, (circle[0], circle[1]), 2, (0, 0, 255), 3)
                ax[j + 1].imshow(imgj, aspect='auto')
                ax[j + 1].set_xticks([])
                ax[j + 1].set_yticks([])
            plt.show()
    except Exception as e:
        print(f"Exception occurred for image {path}: {e}")
        f, ax = plt.subplots(1, 1, figsize=(3, 3))
        ax.imshow(img, aspect='auto')
        ax.set_xticks([])
        ax.set_yticks([])
        plt.show()
        continue
```
